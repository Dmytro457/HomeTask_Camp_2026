## Завдання про суму та добуток 

Є два цілих числа: 1 < a, b <100
 
Одному вченому "Сума" надано суму чисел,
іншому "Добуток" — добуток чисел. 
Обидва вчені знають, що 1 < a, b < 100.

Потрібно визначити числа, спираючись на наведений нижче діалог: 

Добуток: Я не знаю цифр;   

Сума: А я знаю; 

Добуток: Тоді я теж знаю цифри;

Сума: Тоді я теж знаю цифри.

### Step 1: Добуток: Я не знаю цифр; 

Він може не знати ці числа тільки якщо, для його числа є більше чим 2 розклади добутку, і він не може бути впевнений який з них правильний

```python
products_with_multiple = set()
for x in range(2, 100):
    for y in range(x, 100):
        product = x * y
        number_of_pairs = 0

        for a in range(2, 100):
            for b in range(a, 100):
                if a * b == product:
                    number_of_pairs += 1

        if number_of_pairs >= 2: # більш чим 2 пари добутку числ
            products_with_multiple.add(product)
        if(len(products_with_multiple) < 10): # Щоб показати перших 10 числ
            print(f' -> {product}') 
       

print(f'Кількість добутків: {len(products_with_multiple)} з більш чим 2 можливих розкладів')
```

### Step 2: Сума: Я знаю це; 

А він міг вже зрозуміти це тільки якщо, його всі розклади пари сум числ, в добутку дають числа які відповідають умовам з 1 кроку,
тому він знав, що він точно не міг знати числа

```python
sum_numbers = []
for sum_ in range(4, 200):
    total_pairs = 0
    unique_pairs = 0

    for a in range(2, 100):
        for b in range(2, 100):
            if a + b == sum_:
                total_pairs += 1

                for product_ in products_with_multiple:
                    if a * b == product_:
                        unique_pairs += 1

    if unique_pairs == total_pairs: # щоб пари добутку числ, була така сама кількість як і пар сум числ
        sum_numbers.append(sum_)
                                                               
print(f'Суми де всі розклади є неоднозначні: {sum_numbers}')

```

### Step 3: Добуток: Тоді я теж знаю цифри;

Тоді він може знати числа якщо, є тільки одна пара, яка в сумі дає число яке відповідає умовам з 2 кроку   


```python
pairs_for_condition = []

for q in products_with_multiple:
     valid_pairs = []

     for j in range(2, 100):
        for g in range(2, 100):
            if j * g == q:

                for h in sum_numbers:
                    if j + g == h:
                        valid_pairs.append((j, g))

     if len(valid_pairs) == 2: # враховуючи додаткову комбінацію пар числ
        pairs_for_condition.append(valid_pairs[0])
        if len(pairs_for_condition) < 10: # Щоб показати перших 10 пар
            print(" -> ", pairs_for_condition) 
     
 
print(f'Кількість пар які підходять за умовою: {len(pairs_for_condition)}')

```

### Step 4: Сума: Тоді я теж знаю цифри.

Тоді він знає число якщо, в одній із пар сума дає число що відповідає 2 кроку, і є в одній із пар в 3 кроку  


```python
final_pairs = []
for h in sum_numbers:
    valid_pairs = []
    
    for j in range(2, 100):
            for g in range(2, 100):
                if j + g == h:
                    if (j, g) in pairs_for_condition:
                        valid_pairs.append((j, g))       
                          
    if len(valid_pairs) == 1: # тільки 1 можлива пара
        final_pairs.append(valid_pairs[0])     

final_pairs[0]
                        
```

### Відповідь на завдання

```python
print(f'Отже спираючись на їхній діалог було визначено пару числ: {final_pairs[0]}')
print('А отже:')
print(f'Вчений "Добуток" мав число добуток: {final_pairs[0][0] * final_pairs[0][1]}')
print(f'А вчений "Сума" мав число суму: {final_pairs[0][0] + final_pairs[0][1]}')
```
